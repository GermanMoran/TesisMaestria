#### METODO MERGE
ref: https://towardsdatascience.com/python-pandas-tricks-3-best-methods-4a909843f5bc#:~:text=To%20use%20merge()%20%2C%20you,at%20least%20below%20two%20arguments.&text=for%20example%2C%20combining%20above%20two,to%20combine%20the%20two%20datasets.&text=Image%20by%20Author-,pd.,combines%20them%20on%20this%20column.

- *pd.merge(df1,df2)* :Devuelve un conjunto de datos final que contiene solo las filas comunes de ambos conjuntos de datos. Por lo tanto, esto da como resultado una inner JOIN
- *pd.merge(df1, df2, how='outer')* : El 'Outer' es similar a la realizada en SQL. Devuelve filas coincidentes de ambos conjuntos de datos más filas que no coinciden.
- *pd.merge(df1, df2, how='left')* :  left join devuelve todas las filas del DataFrame izquierdo y solo las filas coincidentes del DataFrame derecho
- *pd.merge(df1, df2, how='rigth')* :  rigth join devuelve todas las filas del DataFrame derecho y solo las filas coincidentes del DataFrame derecho
- ️El orden de las columnas en el resultado final cambiará según el orden en que mencione DataFrames en pd.merge()
