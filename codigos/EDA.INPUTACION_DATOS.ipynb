## IMPUTACION DE DATOS

**En este caso como el porcentaje de valores nulos es menor al 1% se decidio utilizar tecnicas (moda , mediana, variznza )estadisticas para realizar dicha emputación**

```python
# Librerias y dependencias
import pandas as pd
from statistics import median,mode
```

```python
# Lectura del archivo
matrizFinal = pd.read_excel("../Archivos Generados/matriz_final.xlsx")
```

```python
# Muestra del dataframe
matrizFinal.head(5)
```

```python
# Descrincion General de las columnas del datset
matrizFinal.info()
```

```python
# conteo de la cantidad de nuelos  en cada una de las columnas
matrizFinal.isnull().sum().sort_values(ascending=False).head(20)
```

```python
# Para material Genetico
matrizFinal.MATERIAL_GENETICO.value_counts()
```

```python
# Se calcula moda para la caracteristica MATERIAL_GENETICO
mode_MG = mode(matrizFinal.MATERIAL_GENETICO)
print("Moda Material Gnetico: ", mode_MG)
```

```python
# Variable Categorica Tipo de Siembra
mode_TS = mode(matrizFinal.TIPO_SIEMBRA)
print("La moda para la variable TIPO_SIEMBRA: ",mode_TS)
```

```python
# Variable numerica | DIAS_EN_EMERGER
median_DEE = median(matrizFinal.DIAS_EN_EMERGER)
print("La media parala variable DIAS_EN_EMERGER es: ",median_DEE)
```

```python

# Variable numerica | DIAS_EN_EMERGER_A_FLORECER
median_DEEF = median(matrizFinal.DIAS_EN_EMERGER_A_FLORECER)
print("La media parala variable DIAS_EN_EMERGER es: ",median_DEEF)
```

```python

# Variable numerica | DIAS_EN_FLORECER_A_COSECHAR
median_DEFC = median(matrizFinal.DIAS_EN_FLORECER_A_COSECHAR)
print("La media parala variable DIAS_EN_FLORECER_A_COSECHAR es: ",median_DEFC)
```

```python
# Variable categorica  | CULT_ANT
mode_CA = mode(matrizFinal.CULT_ANT)
print("La moda parala variable CULT_ANT es: ",mode_CA)
```

```python
# variable numerica  | PROFUND_MOTEADOS_RASTA  
median_PMR =median(matrizFinal.PROFUND_MOTEADOS_RASTA)
print("La media para al variable PROFUND_MOTEADOS_RASTA es: ", median_PMR)
```

```python
matrizFinal[matrizFinal.PROFUND_MOTEADOS_RASTA.isnull()]["PROFUND_MOTEADOS_RASTA"]
```

```python
# Se establece los valores de c/d metrica al dataset final.
matrizFinal.MATERIAL_GENETICO[matrizFinal.MATERIAL_GENETICO.isnull()] = mode_MG
matrizFinal.TIPO_SIEMBRA[matrizFinal.TIPO_SIEMBRA.isnull()] = mode_TS
matrizFinal.DIAS_EN_EMERGER[matrizFinal.DIAS_EN_EMERGER.isnull()] = median_DEE
matrizFinal.DIAS_EN_EMERGER_A_FLORECER[matrizFinal.DIAS_EN_EMERGER_A_FLORECER.isnull()] = median_DEEF
matrizFinal.DIAS_EN_FLORECER_A_COSECHAR[matrizFinal.DIAS_EN_FLORECER_A_COSECHAR.isnull()] = median_DEFC
matrizFinal.CULT_ANT[matrizFinal.CULT_ANT.isnull()] = mode_CA
matrizFinal.PROFUND_MOTEADOS_RASTA[matrizFinal.PROFUND_MOTEADOS_RASTA.isnull()]= median_PMR
```

```python
# Finalmente se crea un nuevo dataset , despues de realizar la imputación de variables
```

## Tratamiento para el CLIMA

```python
# Lectura de los datos.
clima = pd.read_csv("../Datos/Weather_events.csv")
```

```python
# Columnas del datset Clima
clima.columns
```

```python
# Se realiza un mergue entre la matriz Final y el clima
matrizFinal = pd.merge(matrizFinal,clima,how = 'left', on=["ID_EVENTO"])
```

```python
matrizFinal
```

```python
# Se elimina los registros de eventos sin registros climaticos asociados.
matrizFinal = matrizFinal[(matrizFinal.Temp_Avg_Veg.isnull()== False) & (matrizFinal.Temp_Avg_For.isnull() == False) & (matrizFinal.Temp_Avg_Mad.isnull()==False)]
```

```python
# Se eliminan de la matriz Final las caracteristicas con alta varianza
matrizFinal = matrizFinal.drop(borrarPorVarianza,axis=1)
```
