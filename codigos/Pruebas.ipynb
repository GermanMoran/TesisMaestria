```python
# evaluate an lasso regression model on the dataset
from numpy import mean
from numpy import std
from numpy import absolute
from pandas import read_csv
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import Lasso
# load the dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.csv'
dataframe = read_csv(url, header=None)
data = dataframe.values
X, y = data[:, :-1], data[:, -1]
```

```python
dataframe
```

```python
# y: Valor a predecir 
# X: Matriz  con todas las caracteristicas numericas de los algoritmos

y
```

```python
# Creación del modelo lasso
# define model
model = Lasso(alpha=1.0)
# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force scores to be positive
scores = absolute(scores)
print(scores)
print('Mean MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
# Utilizar el modelo generado por la regresion lasso y hacer prediccion de nuevos datos
# Como en la validacion cruzada se utilizan todos los datos, entonces para realizar la prediccion se utilizan nuevos registros

# fit model
model.fit(X, y)
# define new data
row = [0.00632,18.00,2.310,0,0.5380,6.5750,65.20,4.0900,1,296.0,15.30,396.90,4.98]
# make a prediction
yhat = model.predict([row])
# summarize prediction
print('Predicted: %.3f' % yhat)
```

```python
# Atributos
# Muestra los coeficientes seleccionados del modelo
print(model.coef_)
# Numero de caracteristicas tomadas en el ajuste
print(model.n_features_in_)
# Nombre de las caracteristicas del ajuste
#https://machinelearningmastery.com/lasso-regression-with-python/
print(model.feature_names_in)
```

# Ajuste de Hiperparametros del modelo lasso
- El hiperparametro mas importante para ajustar en el algoritmo es  (alfa ) que hace referencia a la tasa de regularizacón L 
-  Se utiliza el enfoque de cuadricula para obtener la mejor configuración de parametros
- El enfoque sería probar valores entre 0,0 y 1,0 con una separación de cuadrícula de 0,01.
- Sklearn proporciona  una clases llamada LassoCV class. en donde se cencuentra de manera automatica lso mejores parametros y se pude simplemente ese modelo ajustado utilizar para realizar las predicciones.


```python
# grid search hyperparameters for lasso regression
from numpy import arange
from pandas import read_csv
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import Lasso
# load the dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.csv'
dataframe = read_csv(url, header=None)
data = dataframe.values
X, y = data[:, :-1], data[:, -1]
# define model
model = Lasso()
# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# define grid
grid = dict()
grid['alpha'] = arange(0, 1, 0.01)
# define search
search = GridSearchCV(model, grid, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# perform the search
results = search.fit(X, y)
# summarize
print('MAE: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)
```

# Implmentación de la libreria  LassoCv

```python
# use automatically configured the lasso regression algorithm
from numpy import arange
from pandas import read_csv
from sklearn.linear_model import LassoCV
from sklearn.model_selection import RepeatedKFold
# load the dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.csv'
dataframe = read_csv(url, header=None)
data = dataframe.values
X, y = data[:, :-1], data[:, -1]
# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# define model
model = LassoCV(alphas=arange(0, 1, 0.01), cv=cv, n_jobs=-1)
# fit model
model.fit(X, y)
```

```python
# Se muestra la mejor configuración seleccionada
print('alpha: %f' % model.alpha_)
```
