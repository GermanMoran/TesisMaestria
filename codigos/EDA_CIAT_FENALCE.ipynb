##### Teniendo en cuenta el archivo production_events_3 las variables que seran eliminadas por falta de datos son las siguientes:

* TIPO_DE_SEMILLA
* HABITO_CRECIMIENTO
* ORIGEN_SEMILLA
* INOCULACION_SEMILLAS
* NUEVA_INOCULACION_SEMILLAS
* PRODUCTO_USADO
* NUEVO_MATERIAL_GENETICO
* OTRO_CULT_ANT 
* RESIEMBRA
* OBSERVACIONES_COSECHA

```python
# Se eliminan las columnas del dataset original.
ColumnasEliminar = ['TIPO_DE_SEMILLA','HABITO_CRECIMIENTO','ORIGEN_SEMILLA','INOCULACION_SEMILLAS','NUEVA_INOCULACION_SEMILLAS','PRODUCTO_USADO','NUEVO_MATERIAL_GENETICO','OTRO_CULT_ANT','RESIEMBRA','OBSERVACIONES_COSECHA']
df_ProductionsEvents = Prductions_Events.drop(ColumnasEliminar, axis=1)
print("Cantidad de registros del dataset: ",df_ProductionsEvents.shape)
df_ProductionsEvents.head(5)
```

```python
# Teniendo en cuenta que en el datset se tienen diferentes tipos del producto cosechaso, solamente se hace enfasis en el
# tipo Grano seco.
# Categorias de la variable categorica producto cosechado.
df_ProductionsEvents.PROD_COSECHADO.value_counts()
```

```python
# Como son variables categoricas se utilizan diagramas de caja, o diagramas de violin para poder determinar la relacion
# Entre estas variables
plt.figure(figsize=(10,6))
sns.boxplot(x = "PROD_COSECHADO", y="RDT", data = df_ProductionsEvents)
texto_del_titulo = "Diagrama de cajas de RDT vs PRODUCCION COSECHADA"
plt.xlabel("TIPO DE PRODUCCION")
plt.ylabel("RDT")
plt.title(texto_del_titulo)
```

**Nota** Se puede observar la presencia de algunos valores atipicos asociados la categoria granos seco, teniendoe n cuenta la variable dependiente (RDT: Rendimiento del cultivo de maiz)

```python
# Se filtra el dataset por el tipo de producción grano seco y se  elimina lo demas (ensilajes y otros)
df1_ProductionsEvents = df_ProductionsEvents[df_ProductionsEvents['PROD_COSECHADO'] == 'Grano seco'].reset_index(drop=True)
print("Logintud del datset: ", df1_ProductionsEvents.shape)
df1_ProductionsEvents.head(5)
```
